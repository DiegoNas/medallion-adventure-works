

# Gold – Modelagem Analítica (Star Schema)

## Objetivo
Este notebook cria a camada **Gold**, responsável pela modelagem analítica
dos dados no formato **Star Schema**, otimizando consultas e análises
em ferramentas como Power BI.

A partir dos dados tratados na Silver, são construídas **tabelas fato e dimensões**
com foco em desempenho e facilidade de uso.

## O que é feito aqui
- Criação de tabelas dimensão
- Criação de tabelas fato
- Relacionamentos no modelo estrela
- Otimização dos dados para consumo analítico
- Persistência final na camada Gold

## Arquitetura
- **Fonte:** Lakehouse – Camada Silver
- **Processo:** Modelagem dimensional
- **Destino:** Lakehouse – Camada Gold (Delta)










```sql


CREATE TABLE IF NOT EXISTS projeto_adventure_works.gold.dim_product
USING DELTA
AS
SELECT
    ProductID          AS product_key,
    product_name,
    ProductNumber,
    Color,
    StandardCost,
    ListPrice,
    Size,
    Weight
FROM projeto_adventure_works.silver.production_product;

```

```sql
CREATE TABLE IF NOT EXISTS projeto_adventure_works.gold.dim_customer
USING DELTA
AS
SELECT
    CustomerID   AS customer_key,
    PersonID,
    StoreID,
    TerritoryID
FROM projeto_adventure_works.silver.customer;
```

```sql
CREATE TABLE IF NOT EXISTS projeto_adventure_works.gold.fact_sales
USING DELTA
AS
SELECT
    soi.SalesOrderID         AS sales_order_key,
    so.OrderDate             AS date_key,
    so.CustomerID            AS customer_key,
    soi.ProductID            AS product_key,
    soi.OrderQty,
    soi.UnitPrice,
    soi.UnitPriceDiscount,
    soi.LineTotal            AS revenue
FROM projeto_adventure_works.silver.sales_order so
JOIN projeto_adventure_works.silver.sales_order_item soi
    ON so.SalesOrderID = soi.SalesOrderID;
```
